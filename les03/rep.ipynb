### Задание №1. Создать массив 10x10 со случайными значениями, найти минимум и максимум (низкая)

```python
import numpy as np

arr = np.random.randint(-1000, 1000, size=(10,10))

print(arr)

print("\nMin - " + str(np.min(arr)))
print("Max - " + str(np.max(arr)))
```

### Задание №2. Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке

```python
import numpy as np

size = 8

arr = np.zeros((size,size))

for i in range(size):
  for j in range(size):
    if i % 2 != 0 and j % 2 != 0:
      continue
    if i % 2 != 0 or j % 2 != 0:
      arr[i,j] = 1

print(arr)
```

### Задание №3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

```python
import numpy as np

z = np.arange(11)

print("Исходный массив:" + str(z))

for i in range(z.size):
  if z[i] > 3 and z[i] < 8:
    z[i] *= -1

print("Итоговый массив:" + str(z))
```

### Задание №4. Найти среднее заначение строки матрицы и вычесть его из каждого елемента строки

```python
import numpy as np

z = np.random.rand(5, 10)
average = np.mean(z, axis=1, keepdims=True)

print("Исходный массив\n" + str(z))
print("\nСреднее арифметическое\n" + str(average) + "\n")

print("Итоговый массив: \n" + str(np.subtract(z, average)))
```

### Задание №5. Поменяйте 1 и 2 строки матрицы

```python
import numpy as np

z = np.arange(25).reshape(5,5)
print("Исходное:\n" + str(z))

tmp = z[0][:].copy()
z[0][:] = z[1][:]
z[1][:] = tmp

print("\nИтог:\n" + str(z))
```

### Задание №6. Дан массив 10x2 подеставлющий собой точки в декартовой системе координат, преобразовать в полярную

```python
import numpy as np
import math

z = np.random.random((10,2))
res = z.copy()

res[:, 0] = np.sqrt(np.square(z[:, 0]) + np.square(z[:, 1]))
res[:, 1] = np.arctan2(z[:, 1], z[:, 0])

print("Исходное:\n", z)
print("\nРезультат (r, theta):\n", res)
```

### Задание №7. Заменить максимальный элемент массива на ноль

```python
import numpy as np

z = np.random.random(10)
max = z.max()

print("Максимум:", max)
print("\nИсходное:\n", z)

z[(np.where(z == max))] = 0

print("\nРезультат:\n", z)
```

### Задание №8. Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз

```python
import numpy as np

z = np.ones(10)
i = np.random.randint(0,len(z),20)
chk = ""

print("Первый массив:\n", z)
print("\nВторой массив:\n", i)

for char in i:
  if str(char) in chk:
    continue
  else:
    z[char] += 1
    chk += str(char)

print("\nРезультат:\n", z)
```

### Задание №10. Найти ближайшее к заданному значению число в заданном массиве

```python
import numpy as np

z = np.arange(100)
v = np.random.uniform(0,100)
chk = v - int(v)

print("Массив:\n", z)
print("\nИсходное число:\n", v)

print((z[int(v)], z[int(v) + 1])[chk > 0.5])
```

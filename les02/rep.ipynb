### Задание №1. https://leetcode.com/problems/two-sum/

```python
class Solution1(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        # Ваше решение

class Solution(object):
    def twoSum(self, nums, target):
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[j] == target - nums[i]:
                    return [i, j]
```

### Задание №2. https://leetcode.com/problems/palindrome-number/description/

```python
class Solution2(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """

class Solution(object):
  def isPalindrome(self, x):
    if x < 0:
      return False
    y=str(x)
    rev = ''.join(reversed(y))
    if rev == y:
      return True
    else:
      return False
```

### Задание №3. https://leetcode.com/problems/roman-to-integer/description/

```python
class Solution3(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """

class Solution(object):
    def romanToInt(self, s):
        roman_map = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        total = 0
        for i in range(len(s) - 1):
            if roman_map[s[i]] < roman_map[s[i+1]]:
                total -= roman_map[s[i]]
            else:
                total += roman_map[s[i]]
        total += roman_map[s[-1]]
        return total
```

### Задание №4. https://leetcode.com/problems/longest-common-prefix/description/

```python
class Solution4(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """

class Solution(object):
    def longestCommonPrefix(self, strs):
        if not strs:
            return ""

        shortest = min(strs, key=len)

        for i, char in enumerate(shortest):
            for other in strs:
                if other[i] != char:
                    return shortest[:i]

        return shortest
```

### Задание №5. https://leetcode.com/problems/merge-two-sorted-lists/description/

```python
class Solution5(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

class Solution(object):
    def mergeTwoLists(self, list1, list2):
        a = []
        while list1:
            a.append(list1.val)
            list1 = list1.next
        while list2:
            a.append(list2.val)
            list2 = list2.next
        a.sort()

        dummy = ListNode()
        curr = dummy
        for x in a:
            curr.next = ListNode(x)
            curr = curr.next

        return dummy.next
```

### Задание №6. https://leetcode.com/problems/longest-substring-without-repeating-characters/description/

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """

class Solution(object):
    def lengthOfLongestSubstring(self, s):
        a = []
        prev = ""

        for i in s:
            if i in prev:
                prev = prev[prev.index(i)+1:]
            prev += i
            a.append(len(prev))

        return max(a) if a else 0
```
